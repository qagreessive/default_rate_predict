```python
import pandas as pd
import numpy as np
```

```python
# 读取测试集数据和训练集数据
test=pd.read_csv('testA.csv')
train=pd.read_csv('train.csv')
```

```python
# 确定差别只是是否违约以后，先对数据集与测试集进行融合，方便后续对缺失值、数据类型的转换处理
data=pd.concat([train,test],axis=0,ignore_index=True)
data
```

```python
# 查看缺失值
missing=data.isnull().sum()
missing_rate=missing/len(data)
missing_rate
```

```python
# 查看data表格中数值小于10的数，发现policyCode列只有一个值，该列可以删掉对后续模型没有影响
u1=data.nunique()
u1[u1<10]
```

```python
del data['policyCode']
data
```

```python
# 筛选出object类型的列
data.loc[:,data.dtypes==object]
```

```python
# 先对grade列、subGrade列进行处理
print(data['grade'].nunique())
print(data['subGrade'].nunique())
```

```python
data=pd.get_dummies(data,columns=['grade','subGrade'],drop_first=True)
data
```

```python
data
```

```python
# 再对employmentLength进行处理
data['employmentLength'].value_counts(dropna=False).sort_index()
```

```python
data['employmentLength'].replace('< 1 year','0 year',inplace=True)
data['employmentLength'].replace('10+ years','10 year',inplace=True)
```

```python
data['employmentLength'].value_counts(dropna=False).sort_index()
```

```python
def employmentLength_to_int(s):
    if pd.isnull(s):
        return s
    else:
        return np.int8(s.split()[0])
```

```python
data['employmentLength']=data['employmentLength'].apply(employmentLength_to_int)
data['employmentLength']
```

```python
# 再对issueDate、earliesCreditLine进行处理
data['issueDate']=data['issueDate'].apply(lambda s1:int(s1[:4]))
data['earliesCreditLine']=data['earliesCreditLine'].apply(lambda s2:int(s2[-4:]))
data.loc[:,['issueDate','earliesCreditLine']]
```

```python
data.info()
```

#### 'employmentTitle','postCode','title' groupby([f])['id']
#### 'homeOwnership', 'verificationStatus','purpose', 'regionCode' get_dummies

```python
data.loc[:,['employmentTitle','postCode','title','homeOwnership', 'verificationStatus','purpose', 'regionCode']]
```

```python
data
```

```python
data=pd.get_dummies(data,columns=['homeOwnership','verificationStatus','purpose','regionCode'],drop_first=True)
data
```

```python
# 网上的做法
# 高维类别特征需要进行转换
for f in ['employmentTitle', 'postCode', 'title']:
    data[f+'_cnts'] = data.groupby([f])['id'].transform('count')
    data[f+'_rank'] = data.groupby([f])['id'].rank(ascending=False).astype(int)
    del data[f]
```
